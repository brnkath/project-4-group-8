```python
# Import the modules
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
path = "Resources/emails.csv"
email_df = pd.read_csv(path)
```

```python
# Drop the Email No. column
email_df = email_df.drop(columns=['Email No.','Prediction'])
```

```python
# Narrow the columns to only words that are over 6 letters
email_df_filtered = [col for col in email_df.columns if len(col) > 3]
email_df = email_df[email_df_filtered]
```

```python
# Extract column names (header words)
header_words = list(email_df.columns)
```

```python
# Calculate the sum of each column
column_sums = email_df.sum()
```

```python
# Map column sums to the corresponding header words
word_sizes = {word: column_sums[word] for word in header_words}
```

```python
# Generate word cloud using word_sizes as frequencies
wc_banner = WordCloud(width=1200, height=200, background_color='black').generate_from_frequencies(word_sizes)
```

```python
# Plot the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wc_banner, interpolation='bilinear')
plt.axis('off')
plt.savefig('img/wordcloud_banner.png', transparent=True)
plt.show()
```

```python
# Generate word cloud using word_sizes as frequencies
wc_4_by_6 = WordCloud(width=600, height=400, background_color='white').generate_from_frequencies(word_sizes)
```

```python
# Plot the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wc_4_by_6, interpolation='bilinear')
plt.axis('off')
plt.savefig('img/wordcloud_medium.png', transparent=True)
plt.show()
```
